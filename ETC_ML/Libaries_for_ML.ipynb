### Encoding

```python
insurance.head()
```

```python
insurance['smoker_region'] = np.where(insurance.smoker == 'yes',1,0)
```

```python
insurance.head()
```

```python
mapping_dict = {'female':1, 'male': 0}
insurance.sex.map(mapping_dict)
```

```python
insurance['sex_le'] = insurance.sex.map(mapping_dict)
insurance.head()
```

```python
insurance.charges = insurance.charges.map(lambda val: round(val,2))
```

```python
insurance.head()
```

```python
insurance['write_mess'] = insurance.apply(lambda row: 'sm' if (row['smoker']=='yes') & (row['children']>3) else "", axis=1)
```

```python
insurance.head()
```

```python
insurance.write_mess.value_counts()
```

```python
insurance.loc[insurance.write_mess == 'sm',:]
```
